### Baseline Accuracy 
#### 38.9% for the majority class

```python
# Defining X and y
X = dummy_df.drop(columns=['date', 'fire_bins'])
y = dummy_df['fire_bins']

# Getting baseline accuracy
y.value_counts(normalize=True)
```

---

### Train Test Split

```python
# Instantiating a train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size = 0.2)

# Scaling data
ss = StandardScaler()
X_train_sc = ss.fit_transform(X_train)
X_test_sc = ss.transform(X_test)
```

### Logistic Regression

```python
#Log Reggresion 
log = LogisticRegression(max_iter=5000)

log_params = {
    'fit_intercept': [True, False],
    'C': [1, 0.5, 0.1]
}

logistic_model_gridsearch = GridSearchCV(log, log_params, cv = 5, n_jobs=8)

# Fitting to unscaled data and evaluating performance
logistic_model_gridsearch.fit(X_train, y_train)

print('Logistic Regression Evaluation with Unscaled Data:')
print(f'Best Parameters         = {logistic_model_gridsearch.best_params_}')
print(f'Best Estimator          = {logistic_model_gridsearch.best_estimator_}')
print(f'Best Cross-Value Score  = {logistic_model_gridsearch.best_score_}')
print(f'Training Accuracy       = {logistic_model_gridsearch.score(X_train,y_train)}')
print(f'Testing Accuracy        = {logistic_model_gridsearch.score(X_test,y_test)}')

# Fitting to scaled data and evaluating performance
logistic_model_gridsearch.fit(X_train_sc, y_train)

print('------------------------------------------------------------')
print('Logistic Regression Evaluation with Scaled Data:')
print(f'Best Parameters         = {logistic_model_gridsearch.best_params_}')
print(f'Best Estimator          = {logistic_model_gridsearch.best_estimator_}')
print(f'Best Cross-Value Score  = {logistic_model_gridsearch.best_score_}')
print(f'Training Accuracy       = {logistic_model_gridsearch.score(X_train_sc,y_train)}')
print(f'Testing Accuracy        = {logistic_model_gridsearch.score(X_test_sc,y_test)}')
```

### $k$ Nearest Neighbors

```python
n_neighbors = [x for x in range(3,20,1)]

knn = KNeighborsClassifier()

knn_params = {
    'n_neighbors': n_neighbors,
    'weights': ['uniform', 'distance'],
    'p': [1, 2],
    'metric': ['euclidean', 'manhattan'],
    'leaf_size' : [2, 4, 6, 8, 10, 12]
}

knn_gridsearch = GridSearchCV(knn, param_grid=knn_params, cv=5, n_jobs=8)

knn_gridsearch.fit(X_train_sc, y_train)
print('KNN Evaluation:')
print(f'Best Parameters         = {knn_gridsearch.best_params_}')
print(f'Best Estimator          = {knn_gridsearch.best_estimator_}')
print(f'Best Cross-Value Score  = {knn_gridsearch.best_score_}')
print(f'Training Accuracy       = {knn_gridsearch.score(X_train_sc,y_train)}')
print(f'Testing Accuracy        = {knn_gridsearch.score(X_test_sc,y_test)}')
```

### Random Forest

```python
# Instantiating random forest model
rf = RandomForestClassifier()

# Defining params to test in our RF model
rf_params = {
    'n_estimators': [5, 10, 25, 50],
    'max_depth': [3, 5, 7, 15],
    'max_features': [None, 'auto', 'log2'],
    'ccp_alpha': [0, 0.01, 0.1]
}

# Instantiating a gridsearch
rf_gs = GridSearchCV(rf, param_grid=rf_params, cv = 5, n_jobs=8)

# Fitting RF model on scaled data
rf_gs.fit(X_train_sc, y_train)

print('Random Forest Evaluation with Scaled Data:')
print(f'Best Params       = {rf_gs.best_params_}')
print(f'Best Score        = {rf_gs.best_score_}')
print(f'Training Accuracy = {rf_gs.score(X_train_sc, y_train)}')
print(f'Testing Accuracy  = {rf_gs.score(X_test_sc, y_test)}')

# Instantiating a gridsearch
rf_gs_us = GridSearchCV(rf, param_grid=rf_params, cv = 5, n_jobs=8)

# Fitting RF model on unscaled data
rf_gs_us.fit(X_train, y_train)

print('------------------------------------------------------------')
print('Random Forest Evaluation with Unscaled Data:')
print(f'Best Params       = {rf_gs_us.best_params_}')
print(f'Best Score        = {rf_gs_us.best_score_}')
print(f'Training Accuracy = {rf_gs_us.score(X_train, y_train)}')
print(f'Testing Accuracy  = {rf_gs_us.score(X_test, y_test)}')
```

### Kernel Support Vector Classifier

```python
# Instantiating SVC.
svc = SVC()

svc_params = {
    'C':np.linspace(0.001, 1, 20),
    'kernel':['rbf', 'polynomial', 'sigmoid'],
    'degree':[1, 2, 3]
}

svc_gs = GridSearchCV(estimator=svc, param_grid=svc_params, cv=5, n_jobs=8)

# Fitting SVC model on scaled data
svc_gs.fit(X_train_sc, y_train)
print('Support Vector Classifier Evaluation with Scaled Data:')
print(f'Best Params       = {svc_gs.best_params_}')
print(f'Best Score        = {svc_gs.best_score_}')
print(f'Training Accuracy = {svc_gs.score(X_train_sc, y_train)}')
print(f'Testing Accuracy  = {svc_gs.score(X_test_sc, y_test)}')

# Fitting SVC model on unscaled data
svc_gs_us = GridSearchCV(estimator=svc, param_grid=svc_params, cv=5, n_jobs=8)
svc_gs_us.fit(X_train, y_train)

print('------------------------------------------------------------')
print('Support Vector Classifier Evaluation with Unscaled Data:')
print(f'Best Params       = {svc_gs_us.best_params_}')
print(f'Best Score        = {svc_gs_us.best_score_}')
print(f'Training Accuracy = {svc_gs_us.score(X_train, y_train)}')
print(f'Testing Accuracy  = {svc_gs_us.score(X_test, y_test)}')
```

### Neural Network

```python
# Defining X and y
X = dummy_df.drop(columns=['date', 'fire_bins'])
y = dummy_df['fire_bins']
```

```python
# Reshaping y values for our NN model
y = to_categorical(y)
```

```python
# Instantiating a train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size = 0.2)

# Scaling data
ss = StandardScaler()
X_train_sc = ss.fit_transform(X_train)
X_test_sc = ss.transform(X_test)
```

```python
nn = Sequential()

# Input and first hidden layer
nn.add(Dense(128, input_shape=(X_train_sc.shape[1], ), activation='relu'))
nn.add(Dropout(0.2))

# Hidden layers
nn.add(Dense(64, activation='relu'))
nn.add(Dropout(0.2))
nn.add(Dense(32, activation='relu'))
nn.add(Dropout(0.2))
nn.add(Dense(32, activation='relu'))
nn.add(Dropout(0.2))

# Output layer
nn.add(Dense(6, activation='softmax'))

nn.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Instantiating early stopping callback with a patience of 3
es = EarlyStopping(patience=3)

# Defining history as fit NN model
hist = nn.fit(X_train_sc, y_train,
        validation_data=(X_test_sc, y_test),
        epochs=50,
        batch_size=256,
        callbacks=es,
        verbose=0)
```

```python
# Reference: Lesson 7.02
score = nn.evaluate(X_test_sc, y_test)
labels = nn.metrics_names

print(f'{labels[0].capitalize()} = {score[0]}')
print(f'{labels[1].capitalize()} = {score[1]}')
```

```python
nn.summary()
```

```python
plot_model(nn, show_layer_names=True, show_shapes=True)
```

```python
# Plotting the training and testing accuracy in our NN model
plt.figure(figsize=(13, 9))
plt.plot(hist.history['accuracy'], label = 'Training Accuracy')
plt.plot(hist.history['val_accuracy'], label = 'Testing Accuracy')
plt.title('Neural Network Accuracy', size=18)
plt.xlabel('Epochs', size=15)
plt.ylabel('Accuracy', size=15)
plt.legend();
```

**ABOVE:** This graph depicts the accuracy of our neural network. The training and testing scores suggest that we're overfitting a bit in spite of using an `EarlyStopping` callback, and we're also topping out at about 40% accuracy on our testing data - barely better than the baseline accuracy of 38.9%.
<br>
<br>
**BELOW:** Plotting the loss for our training and testing scores. Again, the divergence of the two, with testing loss remaining relatively high compared to training, implies overfitting. Additionally, the actual loss values for both training and testing data are very high, suggesting this is a poor predictor of our target variable, `fire_bins`.

```python
# Plotting the training and testing loss in our NN model
plt.figure(figsize=(13, 9))
plt.plot(hist.history['loss'], label = 'Training Loss')
plt.plot(hist.history['val_loss'], label = 'Testing Loss')
plt.title('Neural Network Loss', size=18)
plt.xlabel('Epochs', size=15)
plt.ylabel('Loss', size=15)
plt.legend();
```
