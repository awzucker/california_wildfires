### Look at correlation between weather features

```python
df.columns
```

```python
# check how correlated some features are to avoid collinearity
df[['ELEVATION',
       'AWND', 'CLDD', 'DP01', 'DP10', 'DSND', 'DT00', 'DT32', 'DX32', 'DX70',
       'EMNT', 'EMSN', 'EMXP', 'EMXT', 'HTDD', 'PRCP', 'SNOW', 'TAVG', 'TMAX',
       'TMIN']].corr()
```

```python
df[['TAVG', 'TMAX','TMIN']].tail(10)
```

```python
df['TAVG'].value_counts()
```

**TAVG is supposed to be "Computed by adding the unrounded monthly mean maximum (TMAX) and monthly mean minimum (TMIN) temperatures and dividing by 2". However, there are a lot of 0 values. Going to recalculate this.**

```python
# Replace TAVG with actual average of TMAX and TMIN
df['TAVG'] = (df['TMIN'] + df['TMAX']) / 2
df[['TAVG', 'TMAX','TMIN']].tail(10)
```

```python
# Add a new column for 'TRANGE' to show the delta between the avg. high and low temps
df['TRANGE'] = abs(df['TMAX'] - df['TMIN'])
df[['TAVG', 'TMAX','TMIN','TRANGE']].head(10)
```

```python
# Looks as if these columns are switched
df[df['TMAX'] > df['TMIN']]
```

```python
# Swap names for future clarity
df.rename(columns = {'TMAX' : 'TMIN', 'TMIN' : 'TMAX'}, inplace=True)
df[df['TMAX'] > df['TMIN']]
```

```python
df.corr()[['AcresBurned']].sort_values(by='AcresBurned', ascending=False)
```

```python
# checking highest correlations 
df[['ELEVATION',
       'AWND', 'CLDD', 'DP01', 'DP10', 'DSND', 'DT00', 'DT32', 'DX32', 'DX70',
       'EMNT', 'EMSN', 'EMXP', 'EMXT', 'HTDD', 'PRCP', 'SNOW', 'TAVG', 'TMAX',
       'TMIN', 'TRANGE']].corr().unstack().sort_values(ascending= False).drop_duplicates()[:30]
```

```python
# Reference: https://stackoverflow.com/questions/29294983/how-to-calculate-correlation-between-all-columns-and-remove-highly-correlated-on
# Make list of columns with over 80% correlation to drop

# Create correlation matrix
corr_matrix = df[['ELEVATION',
       'AWND', 'CLDD', 'DP01', 'DP10', 'DSND', 'DT00', 'DT32', 'DX32', 'DX70',
       'EMNT', 'EMSN', 'EMXP', 'EMXT', 'HTDD', 'PRCP', 'SNOW', 'TAVG', 'TMAX',
       'TMIN', 'TRANGE']].corr().abs()

# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# Find features with correlation greater than 0.80
to_drop = [column for column in upper.columns if any(upper[column] > 0.80)]

to_drop
```
