**BELOW:** Imputing missing values in `duration` using the ratio of mean acres burned to mean duration.

```python
df['duration'].isnull().sum()
```

```python
# The ratio calculated below will yield the average acres burned per day. We can use this to impute missing values in 'duration'
ratio = df['acres_burned'].mean() / df['duration'].mean()
ratio
```

```python
# Creating a list of indices where all NaNs in duration column are found
duration_nan_index = (df[df['duration'].isna()].index).tolist()
```

```python
# Viewing first 5 rows in duration column that are NaN
df[df['duration'].isna()].head()
```

```python
# Creating a function to replace all nan values with calculated ratio
def impute_duration(indices):
    for n in indices:
        df.at[n, 'duration'] = round((df.at[n, 'acres_burned'] / ratio), 2)
    return df['duration'].describe()
```

```python
impute_duration(duration_nan_index)
```

```python
# Check first 3 rows that had NaNs to see results
print(f"Acres Burned is 91 and imputed duration is now {df.at[669,'duration']} days")
print(f"Acres Burned is 1011 and imputed duration is now {df.at[673,'duration']} days")
print(f"Acres Burned is 100 and imputed duration is now {df.at[678,'duration']} days")
```

**NOTE:** According to the data dictionary, `avg_monthly_temp` is "*Computed by adding the unrounded monthly mean maximum (TMAX) and monthly mean minimum (TMIN) temperatures and dividing by 2*". However, there are a lot of 0 values. Recalculating this below.

```python
# Replacing avg monthly temp with actual average of max_monthly_temp and min_monthly_temp
df['avg_monthly_temp'] = (df['min_monthly_temp'] + df['max_monthly_temp']) / 2
df[['avg_monthly_temp', 'max_monthly_temp', 'min_monthly_temp']].tail(10)
```

---

**BELOW:** Feature engineering new columns for the temperature range, season, and month.

```python
# Add a new column for 'temp_range' to show the delta between the avg high and low temps
df['temp_range'] = abs(df['max_monthly_temp'] - df['min_monthly_temp'])
df[['avg_monthly_temp', 'max_monthly_temp', 'min_monthly_temp', 'temp_range']].head(10)
```

```python
df['date'].head(2)
```

```python
# Building a list of all the months using the date column
month_list = []
for n in df['date']:
    month_list.append(n[5:7])
```

```python
# Creating two new columns both using the month list (we will manipulate the season column next)
df['month'] = month_list
df['season'] = month_list
df.head(2)
```

```python
# Defining our seasons
winter = ['12', '01', '02']
spring = ['03', '04', '05']
summer = ['06', '07', '08']
fall = ['09', '10', '11']
```

```python
# Creating a list of seasons for each row in the the df
season_list = []
for month in df['season']:
    if month in winter:
        month = 'winter'
    elif month in spring:
        month = 'spring'
    elif month in summer:
        month = 'summer'
    else:
        month = 'fall'
    season_list.append(month)
```

```python
# Changing season column to list of seasons
df['season'] = season_list
```

---

**BELOW:** Imputing nulls in `avg_wind_speed` and `dsnd` (days with more than 1 inch of snow) based on the mean wind speed for the corresponding month.

```python
# Looking at average wind speed by month
month_wind = df['avg_wind_speed'].groupby(by=df['month']).mean().to_dict()

new_wind = [month_wind[df['date'][x].split('-')[2]] if pd.isnull(df['avg_wind_speed'][x]) else df['avg_wind_speed'][x] for x in range(len(df['avg_wind_speed']))]
df['avg_wind_speed'] = new_wind
```

```python
# Looking at average days with snowfall >1 inch by month
month_dsnd = df['dsnd'].groupby(by=df['month']).mean().to_dict()

new_dsnd = [month_dsnd[df['date'][x].split('-')[2]] if pd.isnull(df['dsnd'][x]) else df['dsnd'][x] for x in range(len(df['dsnd']))]
df['dsnd'] = new_dsnd
```

```python
# Checking for nulls in avg wind speed and dsnd
print(df['avg_wind_speed'].isna().sum())
print(df['dsnd'].isna().sum())
```

```python
# Reference: https://stackoverflow.com/questions/29294983/how-to-calculate-correlation-between-all-columns-and-remove-highly-correlated-on
# Making a list of columns with over 80% correlation to drop

# Create correlation matrix
corr_matrix = df[['elevation', 'avg_wind_speed', 'cooling_degree_days', 'dp01', 'dp10', 'dsnd', 'dt00', 'dt32', 'dx32', 
    'dx70', 'extreme_min_temp', 'highest_daily_snowfall', 'highest_daily_precipitation', 'extreme_max_temp', 
    'heating_degree_days', 'total_monthly_precipitation', 'total_monthly_snowfall', 'avg_monthly_temp', 
    'max_monthly_temp', 'min_monthly_temp', 'temp_range']].corr().abs()

# Selecting upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# Finding features with correlation greater than 0.80
to_drop = [column for column in upper.columns if any(upper[column] > 0.80)]

to_drop
```
